## Under-Sampling
- Reasons for choosing  Under-Sampling here:
1) Dataset Is Very Large : The dataset contains hundreds of thousands of normal transactions and Removing some normal transactions does not lose important information. So under-sampling is safe here.

2) Fraud Data Is Real and Limited : Fraud transactions are real, rare, and important and we should not duplicate or artificially create fraud
data. So, Under-sampling keeps only real data.

3) Avoid Overfitting(Compared to Oversampling) : Oversampling repeats fraud transactions, repeated data can cause the model to memorize, not learn, Under-Sampling avoids Duplication and So, therefore Logistic Regression stays Generalized.

4) Faster Training : Smaller dataset means faster model training, Less memory usage and Useful when computational resources are limited.

```python
#1) For Under - Sampling, Building a Sample dataset containing similar distribution of Normal Transactions and Fraudulent Transactions.
#2) Number of Fraudulent Transactions --> 473.
```

```python
fraud_count = fraud.shape[0]
fraud_count
```

```python
legit_sample = legit.sample(n=473)
```

```python
#) Concatenating two DataFrames.
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts() 
```

## Splitting the data into Features & Targets.

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Split the data into Training data & Testing Data.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training(Logistic Regression Model).

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

## Model Evaluation

## Accuracy Score

```python
# Accuracy on Training Data.
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training data : ", training_data_accuracy)
```

1) The model was trained using Training Data.
2) Accuracy ≈ 93.9%
3) This means that the model correctly predicted about 94 out of 100 transactions in the training dataset.

```python
# Accuracy on Test Data.
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy score on Test Data : ", test_data_accuracy)
```

1) Test data was not seen by the model before.

2)  Accuracy ≈ 92.6%.
   
3)  This means that the model correctly predicted about 93 out of 100 unseen Transactions.

## Precision

```python
#) Precision on Training Data 
X_train_prediction = model.predict(X_train)
training_data_precision = precision_score(Y_train, X_train_prediction)
print("Precision on Training data : ", training_data_precision)
```

- Training Precision(96.63%):-
- Out of 100 transactions marked as fraud(Predicted Fraud by the Model):
1) 97 are actually fraud
2) Only 3 are mistakes

- Therefore, The model learned very well during training.

```python
# Precision on Test Data
X_test_prediction = model.predict(X_test)
test_data_precision = precision_score(Y_test, X_test_prediction)
print("Precision on Test data : ", test_data_precision)
```

- Test Precision(94.51%):-
- Out of 100 transactions marked as fraud(Predicted Fraud by the Model):
1) 95 are actually fraud.
2) 5 are mistakes.
    
- Therefore, the model still performs very well in real life
